# city와 f4를 기준으로 f5의 평균값을 구한 다음, f5를 기준으로 상위 7개 값을 모두 더해 출력하시오 (소수점 둘째자리까지 출력)
 - 데이터셋 : basic1.csv

```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/빅분기/archive/basic1.csv')
df.head()
```

```python
# city와 f4를 기준으로 f5의 평균값을 구하기
df2 = df.groupby(['city', 'f4'])[['f5']].mean()
df2
```

```python
# dataframe 전환 후 상위 7개 출력
# f5기준으로 상위 7개값 구하기
df2 = df2.reset_index().sort_values('f5', ascending=False).head(7)
df2
```

```python
# 상위 7개 값 모두 더하기(소수점 2번째)
print(round(df2['f5'].sum(), 2))
```
