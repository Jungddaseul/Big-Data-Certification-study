```python
  import numpy as np
  import pandas as pd

  df = pd.read_csv('/content/drive/MyDrive/빅분기/archive/covid-vaccination-vs-death_ratio.csv')
  df.head()
```

# 컬럼설명
- country : 나라이름
- iso_code : 국가별iso코드
- date : 이 데이터가 속한 날짜
- total_vaccinations : 모든 COVID 백신사용 횟수
- people_vaccinated : COVID 백신을 1번 이상 맞은 사람
- people_fully_vaccinated : 전체 백신을 맞은 사람의 수
- New_deaths : 일일 신규사망자 수
- population : 2021년 국가 인구
- ratio : 비율

```python
df.isnull().sum()
```

```python
# 국가별 접종률 구하기(시간이 지날수록 접종률이 높아지므로, 가장 높은 접종률 출력)
df2 = df.groupby('country').max()
df2.head()
```

```python
# 접종률이 높은 순으로 정렬(내림차순)
df2 = df2.sort_values(by='ratio', ascending=False)
df2.head()
```

```python
# 이상치 제거(비율이 100이상인 행 제거)
df2 = df2[3:]
```

```python
# 상위 10%, 하위 10% 구하기(1)
top = df2['ratio'][:10].mean()
bottom = df2['ratio'][-10:].mean()
```
