[작업순서]
1. 라이브러리 및 데이터 불러오기
2. ESFJ 값을 가진 데이터 확인
3. ESFJ데이터를 ISFJ로 대체
4. city가 경기이면서 f4가 ISFJ인 데이터 찾기
5. 위 데이터의 age컬럼의 최댓값 구하기

```python
import numpy as np
import pandas as pd

# 1. 라이브러리 및 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/빅분기/archive/basic1.csv')
df
```

```python
# 2. ESFJ 값을 가진 데이터 확인
df[df['f4'] == 'ESFJ']
```

```python
# 3. ESFJ데이터를 ISFJ로 대체
df['f4'] = df['f4'].replace('ESFJ', 'ISFJ')
df
```

```python
# 값 확인
df[df['f4'] == 'ESFJ']
```

```python
# 값 확인
df[df['f4'] == 'ISFJ']
```

```python
# 4. city가 경기이면서 f4가 ISFJ인 데이터 찾기
df[(df['city'] == '경기') & (df['f4'] == 'ISFJ')]
```

```python
# 5. 위 데이터의 age컬럼의 최댓값 구하기
result = df[(df['city'] == '경기') & (df['f4'] == 'ISFJ')]['age'].max()
```
