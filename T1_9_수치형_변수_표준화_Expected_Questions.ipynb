```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/빅분기/archive/basic1.csv')
df.head()
```

```python
# 'f5'컬럼 표준화
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['f5'] = scaler.fit_transform(df[['f5']])
df.head()
```

```python
# 중앙값 출력
print(df['f5'].median())
```
